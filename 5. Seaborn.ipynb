## Seaborn
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures. Seaborn helps you explore and understand the data.

```python
# Importing
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

**a. Correlation:**  
corr() method calculates the relationship between each column in your data set.

```python
df.corr()
```

**b. Heat Map:**   
To visualize complex data sets and to identify patterns or relationships between variables.

```python
sns.heatmap(df.corr())
```

**c. Join Plot:**  
Relationship between two variables is displayed on a single graph

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

**d. Pair Plot:**   
Displays pairwise relationships between variables in a dataset.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['sex'].value_counts()
```

**f. Distplot**  
To represent the distribution of a univariate set of observations.

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

## Categorical Features

**g. Countplot**  
A countplot is a type of plot in which the data is represented in a bar chart  

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot('day',data=df)
```

**h. Bar Plot**  
Graph that represents a categorical data with rectangular bars, where the height or length of the bar is proportional to the value of the data.

```python
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
sns.barplot(x='tip', y='day', data=df)
```

**i. Box Plot**  
A box plot is a type of chart that displays the distribution of numerical data based on five summary statistics - minimum, first quartile, median, third quartile, and maximum.

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','tip',data=df, palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

**j. Cat Plot**  
To plot different types of categorical plots in one function.

```python
sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=df, kind="box",height=4, aspect=.7);
```

**k. Violin plot**  
Type of visualization that combines the features of a box plot and a kernel density plot.

```python
sns.violinplot('day','tip',data=df, palette='rainbow')
```
